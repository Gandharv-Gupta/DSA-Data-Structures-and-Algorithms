Extent tree class built in our main tutorial so that it takes name and designation in data part of TreeNode class. Now extend print_tree function such that it can print either name tree, designation tree or name and designation tree. As shown above

```python
class gtree:
    def __init__(self,name,designation):
        self.name=name
        self.designation=designation
        self.children=[]
        self.parent=None

    def add_child(self,child):
        child.parent=self
        self.children.append(child)

    def get_level(self):
        level=0
        p=self.parent
        while p:
            level+=1
            p=p.parent
        return level

    def print_name_tree(self):#to print name tree
        level=self.get_level()
        prefix='-'*level*3+'|'

        print(prefix+self.name)
        if self.children:
            for child in self.children:
                child.print_name_tree()

    def print_designation_tree(self):#to print designation tree
        level=self.get_level()
        prefix='-'*level*3+'|'

        print(prefix+self.designation)
        if self.children:
            for child in self.children:
                child.print_designation_tree()

    def print_combined_tree(self):#to preint a combined tree
        level=self.get_level()
        prefix='-'*level*3+'|'

        print(prefix+self.name+'('+self.designation+')')
        if self.children:
            for child in self.children:
                child.print_combined_tree()

#Adding details to the tree

#creating root node:
node_level_0=gtree('Nilupul','CEO')

node_level_1_1=gtree('Chinmay','CTO')
node_level_1_1_1=gtree('Vsihwa','INFRASTRUCTURE HEAD')
node_level_1_1_2=gtree('Amir','APPLICATION HEAD')
node_level_1_1_1.add_child(gtree('Dhawal','CLOUD MANAGER'))
node_level_1_1_1.add_child(gtree('Abhijit','APP MANAGER'))
node_level_1_1.add_child(node_level_1_1_1)
node_level_1_1.add_child(node_level_1_1_2)
node_level_0.add_child(node_level_1_1)
        
node_level_1_2=gtree('Gels','HR HEAD')
node_level_1_2.add_child(gtree('Peter','RECRUITMENT MANAGER'))
node_level_1_2.add_child(gtree('Waqas','POLICY MANAGER'))
node_level_0.add_child(node_level_1_2)


        
```
